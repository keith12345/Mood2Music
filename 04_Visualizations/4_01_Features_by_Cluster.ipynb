```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
speech_df = pd.read_pickle('../Data/labeled_points_tfidf_sent_rm_pca_full_features_12').iloc[:, -15:]
speech_df['hue'] = speech_df.labels == 3
speech_df = speech_df[['speechiness', 'labels', 'hue']]
```

```python
speech_df.groupby('labels').mean()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='labels', y='speechiness', hue='hue', data=speech_df)
plt.suptitle('Cluster Feature Importance', fontsize=26,
             weight='bold', color='#333335', x=0.36, y=1)
plt.title('Average Instrumentalness by Cluster', x=0.284, y=1.03, fontsize=18,
          weight='bold', color='#333335')
plt.xlabel('Clusters', fontsize=18, weight='bold',
           color='#333335', x=0.07, y=-0.15)
plt.xticks([-1], '')
plt.yticks([-1], '')
plt.ylabel('')
sns.despine()
plt.show();
```
