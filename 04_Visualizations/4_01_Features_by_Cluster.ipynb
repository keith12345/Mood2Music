```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('../Data/labeled_points_tfidf_sent_pca_12').iloc[:, -14:]
```

```python
df.head()
```

```python
speech_df = pd.read_pickle('../Data/labeled_points_tfidf_sent_pca_12').iloc[:, -14:]
speech_df['hue'] = speech_df.labels == 9
speech_df = speech_df[['speechiness', 'labels', 'hue']]
```

```python
speech_df.groupby('labels').mean()
```

```python
plt.bar(speech_df.labels, speech_df.groupby('labels').mean())
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='labels', y='speechiness', hue='hue', data=speech_df)
plt.suptitle('Cluster Feature Importance', fontsize=26,
             weight='bold', color='#333335', x=0.36, y=1)
plt.title('Speechiness', x=0.1, y=1.03, fontsize=18,
          weight='bold', color='#333335')
plt.xlabel('Clusters', fontsize=18, weight='bold',
           color='#333335', x=0.07, y=-0.15)
plt.xticks([-1], '')
plt.yticks([-1], '')
plt.ylabel('')
sns.despine()
plt.show();
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='labels', y='danceability', data=df[['danceability', 'labels']])
plt.suptitle('Cluster Feature Importance', fontsize=26,
          weight='bold', color='#333335', x=0.36, y=1)
plt.title('Danceability', x=0.1, y=1.03, fontsize=18,
          weight='bold', color='#333335')
plt.xlabel('Clusters', fontsize=18, weight='bold', color='#333335', x=0.07, y=-0.15)
plt.xticks([-1], '')
plt.yticks([-1], '')
plt.ylabel('')
sns.despine()
plt.show();
```

```python
feats = df.columns.tolist()[:-1]

for feat in feats:
    sns.barplot(x='labels', y=feat, data=df[[feat, 'labels']])
    plt.show();
```
