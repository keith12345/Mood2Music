```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

## Looking at features that do and do not result in overfitting:

```python
speech_df = pd.read_pickle('../Data/labeled_points_tfidf_sent_rm_pca_full_features_12').iloc[:, -15:]
speech_df['hue'] = speech_df.labels == 3
speech_df = speech_df[['speechiness', 'labels', 'hue']]
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='labels', y='speechiness', hue='hue', data=speech_df)
plt.suptitle('Cluster Feature Importance', fontsize=26,
             weight='bold', color='#333335', x=0.36, y=1)
plt.title('Average Speechiness by Cluster', x=0.284, y=1.03, fontsize=18,
          weight='bold', color='#333335')
plt.xlabel('Clusters', fontsize=18, weight='bold',
           color='#333335', x=0.07, y=-0.15)
plt.xticks([-1], '')
plt.yticks([-1], '')
plt.ylabel('')
sns.despine()
plt.show();
```

```python
df = pd.read_pickle('../Data/labeled_points_tfidf_sent_rm_pca_full_features_12').iloc[:, -15:]
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='labels', y='danceability', data=df[['danceability', 'labels']], color='purple', alpha=0.7)
plt.suptitle('Cluster Feature Importance', fontsize=26,
          weight='bold', color='#333335', x=0.36, y=1)
plt.title('Average Danceability by Cluster', x=0.25, y=1.03, fontsize=18,
          weight='bold', color='#333335')
plt.xlabel('Clusters', fontsize=18, weight='bold', color='#333335', x=0.07, y=-0.15)
plt.xticks([-1], '')
plt.yticks([-1], '')
plt.ylabel('')
sns.despine()
plt.show();
```

## Looking at specific songs

```python
df = pd.read_pickle('../Data/labeled_points_tfidf_sent_rm_pca_9')
```

```python
sleepyhead = df[df.song_title == 'sleepyhead'].iloc[:, -10:-1].values[1]
```

```python
plt.figure(figsize=(16, 8))
plt.bar(x=range(9), height=sleepyhead)
plt.suptitle('Understanding Lyric Features - Cluster 1', fontsize=26,
          weight='bold', color='#333335', x=0.38, y=1)
plt.title('Feature Value', x=0.08, y=1.03, fontsize=18,
          weight='bold', color='#333335')
plt.xlabel('Features', fontsize=18, weight='bold', color='#333335', x=0.055, y=-0.15)

xticks = ['TF-IDF Feature', 'Sentiment 1',
          'Sentiment 2', 'Acousticness',
          'Danceability', 'Energy',
          'Loudness', 'Tempo', 'Valence']

plt.xticks(range(10), xticks, weight='bold', fontsize=11, color='#333335')
plt.yticks([-1], '')
plt.ylabel('')
sns.despine()
plt.show();
```

```python
awakening = df[df.song_title == 'awakening'].iloc[:, -10:-1].values[0]
```

```python
plt.figure(figsize=(16, 8))
plt.bar(x=range(9), height=awakening)
plt.suptitle('Understanding Lyric Features - Cluster 1', fontsize=26,
          weight='bold', color='#333335', x=0.38, y=1)
plt.title('Feature Value', x=0.08, y=1.03, fontsize=18,
          weight='bold', color='#333335')
plt.xlabel('Features', fontsize=18, weight='bold', color='#333335', x=0.055, y=-0.15)

xticks = ['TF-IDF Feature', 'Sentiment 1',
          'Sentiment 2', 'Acousticness',
          'Danceability', 'Energy',
          'Loudness', 'Tempo', 'Valence']

plt.xticks(range(10), xticks, weight='bold', fontsize=11, color='#333335')
plt.yticks([-1], '')
plt.ylabel('')
sns.despine()
plt.show();
```

```python
index = np.arange(9)  # the x locations for the groups
width = 0.35       # the width of the bars

fig = plt.figure(figsize=(16, 8), constrained_layout=True)
ax = fig.add_subplot(111)
rects1 = ax.bar(index, awakening, width, color='seagreen', alpha=0.7)

rects2 = ax.bar(index + width, sleepyhead, width, color='purple', alpha=0.1)

# add some
fig.suptitle('Finding Similar Songs', fontsize=26,
          weight='bold', color='#333335', x=0.155, y=1.07)
ax.set_title('Feature Value', x=0.07, y=1.03, fontsize=18,
          weight='bold', color='#333335')
ax.set_xticks(index + width / 2)
ax.set_xticklabels(['TF-IDF Feature', 'Sentiment 1',
                    'Sentiment 2', 'Acousticness',
                    'Danceability', 'Energy',
                    'Loudness', 'Tempo', 'Valence'],
                   weight='bold', fontsize=11, color='#333335')
plt.xlabel('Features', fontsize=18, weight='bold', color='#333335', x=0.045, y=-0.1)

plt.yticks([-1], '')
plt.ylabel('')
sns.despine()

ax.legend( (rects1[0], rects2[0]), ('Awakening - Empire of the sun',
                                    'Passion Pit - Sleepyhead'),
          prop={'size': 18})

plt.show();
```
